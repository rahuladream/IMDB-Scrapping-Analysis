```python
import pandas as pd
from bs4 import BeautifulSoup
from requests import get
from scrapy import Selector
```

```python
url = 'https://www.imdb.com/title/tt0068646/'
```

```python
page = get(url)
```

```python
soup = BeautifulSoup(page.content, 'lxml')
```

```python
title_wrapper = soup.find('div', {"class": "title_wrapper"})
```

```python
title_wrapper.h1.text
```

```python
storyLine = soup.find('div', {'id': 'titleStoryLine'})
```

```python
storyLineInfo = storyLine.find('div', {'class': 'inline canwrap'})
```

```python
sel = Selector(page)
budget = ' '.join(sel.css(".txt-block:contains('Budget')::text").extract()).strip()
gross = ' '.join(sel.css(".txt-block:contains('Gross USA')::text").extract()).strip()
cumulative = ' '.join(sel.css(".txt-block:contains('Cumulative Worldwide')::text").extract()).strip()
runtime = ' '.join(sel.css(".txt-block:contains('Runtime')::text").extract()).strip()
print(f'budget: {budget}\ngross: {gross}\ncumulative: {cumulative}\nRuntime: {runtime}')
```
